### Visualização de dados com a biblioteca Matplotlib

### Importação das bibliotecas necessárias

```python
# Importando as bibliotecas necessárias
import numpy as np  # Biblioteca para operações numéricas
import pandas as pd  # Biblioteca para manipulação de dados
import matplotlib.pyplot as plt  # Biblioteca para visualização de gráficos
```

### Criação do dataframe com dados fictícios

```python
# Dados fictícios de vendas
dados = {
    'Produto': ['Celular', 'Notebook', 'TV', 'Tablet', 'Fones de Ouvido'],
    'Vendas': [120, 75, 90, 50, 110],
    'Receita': [12000, 7500, 9000, 5000, 11000],
    'Ano': [2018, 2019, 2020, 2021, 2022],
    'Lucro': [1000, 800, 1500, 400, 250],
    'Despesas': [600, 500, 700, 300, 150]
}

df = pd.DataFrame(dados)
```

### Gráfico de Barras (Vendas por Produto):

```python
# Neste gráfico de barras, é possível visualizar as vendas de cada produto eletrônico. É útil para comparar as quantidades vendidas entre os produtos.

plt.bar(df['Produto'], df['Vendas'])  # Criação das barras com os dados de Produto e Vendas
plt.xlabel('Produto')  # Definição do rótulo do eixo x como "Produto"
plt.ylabel('Vendas')  # Definição do rótulo do eixo y como "Vendas"
plt.title('Vendas por Produto')  # Definição do título do gráfico como "Vendas por Produto"
plt.show()  # Exibição do gráfico
```

### Gráfico de Linhas (Receita ao longo dos Anos):

```python
# Este gráfico de linhas mostra a variação da receita da empresa de produtos eletrônicos ao longo dos anos. É útil para identificar tendências e padrões.

plt.plot(df['Ano'], df['Receita'], marker='o')  # Criação do gráfico de linhas com os dados de Ano e Receita, marcadores circulares
plt.xlabel('Ano')  # Definição do rótulo do eixo x como "Ano"
plt.ylabel('Receita')  # Definição do rótulo do eixo y como "Receita"
plt.title('Receita ao longo dos Anos')  # Definição do título do gráfico como "Receita ao longo dos Anos"
plt.show()  # Exibição do gráfico
```

### Gráfico de Pizza (Distribuição das Vendas):

```python
# Neste gráfico de pizza, podemos visualizar a distribuição percentual das vendas por produto. É útil para mostrar a proporção de cada produto em relação ao total das vendas.

plt.pie(df['Vendas'], labels=df['Produto'], autopct='%1.1f%%')  # Criação do gráfico de pizza com os dados de Vendas e rótulos dos Produtos, exibindo a porcentagem com uma casa decimal
plt.title('Distribuição das Vendas')  # Definição do título do gráfico como "Distribuição das Vendas"
plt.legend(df['Produto'], title='Produtos', bbox_to_anchor=(1.5, 0.5), loc="center right")  # Adição da legenda com os rótulos dos Produtos, com título "Produtos" e posição ajustada no canto direito
plt.show()  # Exibição do gráfico
```

### Gráfico de Dispersão (Vendas vs. Receita):

```python
# Este gráfico de dispersão mostra a relação entre as vendas e a receita. É útil para identificar se existe uma correlação entre essas variáveis.

plt.scatter(df['Vendas'], df['Receita'])  # Criação do gráfico de dispersão com os dados de Vendas no eixo x e Receita no eixo y
plt.xlabel('Vendas')  # Definição do rótulo do eixo x como "Vendas"
plt.ylabel('Receita')  # Definição do rótulo do eixo y como "Receita"
plt.title('Vendas vs. Receita')  # Definição do título do gráfico como "Vendas vs. Receita"
plt.show()  # Exibição do gráfico
```

### Gráfico de barras empilhadas (Análise de Receita, Lucro e Despesas por Produto):

```python
# Configuração da figura do gráfico
plt.figure(figsize=(10, 6))  # Definição do tamanho da figura do gráfico como 10x6

# Criação dos gráficos de barras empilhadas
plt.bar(df['Produto'], df['Receita'], label='Receita')  # Criação do primeiro gráfico de barras para a Receita
plt.bar(df['Produto'], df['Lucro'], label='Lucro', bottom=df['Receita'])  # Criação do segundo gráfico de barras para o Lucro, empilhado sobre a Receita
plt.bar(df['Produto'], df['Despesas'], label='Despesas', bottom=df['Receita'] + df['Lucro'])  # Criação do terceiro gráfico de barras para as Despesas, empilhado sobre a Receita e o Lucro

# Configuração dos eixos e títulos do gráfico
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x em 45 graus para melhor legibilidade
plt.xlabel('Produto')  # Definição do rótulo do eixo x como "Produto"
plt.ylabel('Valor')  # Definição do rótulo do eixo y como "Valor"
plt.title('Análise de Receita, Lucro e Despesas por Produto')  # Definição do título do gráfico como "Análise de Receita, Lucro e Despesas por Produto"

# Configuração da legenda do gráfico
plt.legend(bbox_to_anchor=(1.2, 0.5), loc="center right")  # Posiciona a legenda no canto direito da figura

# Exibição do gráfico
plt.show()
```

### Gráfico de área (Evolução da Receita, Lucro e Despesas):

```python
# Configuração da figura do gráfico
plt.figure(figsize=(10, 6))  # Definição do tamanho da figura do gráfico como 10x6

# Criação do gráfico de áreas empilhadas
plt.stackplot(df['Ano'], df['Receita'], df['Lucro'], df['Despesas'], labels=['Receita', 'Lucro', 'Despesas'])  # Criação do gráfico de áreas empilhadas com os dados de Receita, Lucro e Despesas

# Configuração dos eixos e títulos do gráfico
plt.xlabel('Ano')  # Definição do rótulo do eixo x como "Ano"
plt.ylabel('Valor')  # Definição do rótulo do eixo y como "Valor"
plt.title('Evolução da Receita, Lucro e Despesas')  # Definição do título do gráfico como "Evolução da Receita, Lucro e Despesas"

# Configuração da legenda do gráfico
plt.legend(loc='upper right')  # Posiciona a legenda no canto superior direito

# Exibição do gráfico
plt.show()
```

### Histograma (Distribuição do Lucro)

```python
# Configuração da figura do gráfico
plt.figure(figsize=(8, 6))  # Definição do tamanho da figura do gráfico como 8x6

# Criação do histograma do Lucro
plt.hist(df['Lucro'], bins='auto', edgecolor='black')  # Criação do histograma utilizando os dados de Lucro

# Configurações dos eixos e título do gráfico
plt.xlabel('Lucro')  # Definição do rótulo do eixo x como "Lucro"
plt.ylabel('Frequência')  # Definição do rótulo do eixo y como "Frequência"
plt.title('Distribuição do Lucro')  # Definição do título do gráfico como "Distribuição do Lucro"

# Adição de informações adicionais no gráfico
plt.text(0.95, 0.95, f"Média: {df['Lucro'].mean():.2f}", ha='right', va='top', transform=plt.gca().transAxes)  # Adição de texto com a média do Lucro no canto superior direito
plt.text(0.95, 0.88, f'Mediana: {df["Lucro"].median():.2f}', ha='right', va='top', transform=plt.gca().transAxes)  # Adição de texto com a mediana do Lucro logo abaixo do texto da média
plt.text(0.95, 0.81, f'Desvio Padrão: {df["Lucro"].std():.2f}', ha='right', va='top', transform=plt.gca().transAxes)  # Adição de texto com o desvio padrão do Lucro logo abaixo do texto da mediana

# Exibição do gráfico
plt.show()
```

### Boxplot:

```python
# Configuração da figura do gráfico
plt.figure(figsize=(8, 6))  # Definição do tamanho da figura do gráfico como 8x6

# Criação do boxplot
plt.boxplot([df['Vendas'], df['Receita'], df['Lucro'], df['Despesas']], labels=['Vendas', 'Receita', 'Lucro', 'Despesas'])  # Criação do boxplot com os dados de Vendas, Receita, Lucro e Despesas

# Configurações dos eixos e título do gráfico
plt.ylabel('Valor')  # Definição do rótulo do eixo y como "Valor"
plt.title('Boxplot das Variáveis')  # Definição do título do gráfico como "Boxplot das Variáveis"

# Exibição do gráfico
plt.show()
```

### Mapa de calor (Correlações):

```python
# Configuração da figura do gráfico
plt.figure(figsize=(8, 6))  # Definição do tamanho da figura do gráfico como 8x6

# Cálculo da matriz de correlação entre as variáveis
heatmap_data = df[['Vendas', 'Receita', 'Lucro', 'Despesas']].corr()  # Seleção das colunas relevantes do dataframe e cálculo da matriz de correlação

# Criação do mapa de calor
heatmap = plt.imshow(heatmap_data, cmap='coolwarm')  # Criação do mapa de calor usando a matriz de correlação e a paleta de cores 'coolwarm'

# Adição da barra de cores
plt.colorbar(heatmap)  # Adição da barra de cores que representa os valores do mapa de calor

# Configurações dos rótulos dos eixos
plt.xticks(range(len(heatmap_data.columns)), heatmap_data.columns, rotation=45)  # Definição dos rótulos dos ticks do eixo x como os nomes das colunas
plt.yticks(range(len(heatmap_data.columns)), heatmap_data.columns)  # Definição dos rótulos dos ticks do eixo y como os nomes das colunas

# Adicionando bordas pretas aos quadrados
for i in range(len(heatmap_data.columns)):  # Loop pelas linhas da matriz de correlação
    for j in range(len(heatmap_data.columns)):  # Loop pelas colunas da matriz de correlação
        plt.gca().add_patch(plt.Rectangle((j - 0.5, i - 0.5), 1, 1, edgecolor='black', linewidth=1, fill=False))  # Adição de retângulos com bordas pretas

# Configuração do título do gráfico
plt.title('Mapa de Calor - Correlação')  # Definição do título do gráfico como "Mapa de Calor - Correlação"

# Exibição do gráfico
plt.show()
```
