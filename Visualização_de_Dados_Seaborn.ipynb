### Visualização de dados com a biblioteca Seaborn

### Importação das bibliotecas necessárias

```python
# Importando as bibliotecas necessárias
import pandas as pd  # Biblioteca para manipulação de dados
import seaborn as sns # Biblioteca para visualização de gráficos
import matplotlib.pyplot as plt  # Biblioteca para visualização de gráficos
```

### Criação do dataframe com dados fictícios

```python
# Criação do dataframe fictício
data = {
    'Paciente': ['P1', 'P2', 'P3', 'P4', 'P5', 'P6'],
    'Idade': [32, 45, 28, 63, 51, 37],
    'Gênero': ['F', 'M', 'M', 'F', 'F', 'F'],
    'Diagnóstico': ['Câncer', 'Fratura', 'Hipertensão', 'Diabetes', 'Fratura', 'Câncer'],
    'Custo': [1500, 2500, 800, 3000, 2000, 1800],
    'Tempo de Internação': [5, 7, 3, 10, 6, 4]
}

df = pd.DataFrame(data)
```

### Gráfico de barras (contagem de diagnósticos):

```python
# Gráfico de barras (contagem de diagnósticos)
plt.figure(figsize=(8, 6))
sns.countplot(x='Diagnóstico', data=df)
plt.xlabel('Diagnóstico')
plt.ylabel('Contagem')
plt.title('Contagem de Diagnósticos')
plt.show()
```

### Gráfico de linhas (tempo de internação)

```python
# Gráfico de linhas (tempo de internação)
plt.figure(figsize=(8, 6))
sns.lineplot(x='Paciente', y='Tempo de Internação', data=df, marker='o')
plt.xlabel('Paciente')
plt.ylabel('Tempo de Internação')
plt.title('Tempo de Internação por Paciente')
plt.show()
```

### Gráfico de setores (distribuição de gênero)

```python
# Gráfico de setores (distribuição de gênero)
plt.figure(figsize=(8, 6))
df['Gênero'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Distribuição de Gênero')
plt.legend()
plt.show()
```

### Gráfico de dispersão (idade x custo)

```python
# Gráfico de dispersão (idade x custo)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Idade', y='Custo', data=df, hue='Diagnóstico')
plt.xlabel('Idade')
plt.ylabel('Custo')
plt.title('Idade vs. Custo')
plt.legend()
plt.show()
```

### Boxplot (custo por diagnóstico)

```python
# Boxplot (custo por diagnóstico)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Diagnóstico', y='Custo', data=df)
plt.xlabel('Diagnóstico')
plt.ylabel('Custo')
plt.title('Custo por Diagnóstico')
plt.show()
```

### Gráfico de barra horizontal (custo por diagnóstico)

```python
# Gráfico de barra horizontal (custo por diagnóstico)
plt.figure(figsize=(8, 6))
sns.barplot(x='Custo', y='Diagnóstico', data=df)
plt.xlabel('Custo')
plt.ylabel('Diagnóstico')
plt.title('Custo por Diagnóstico')
plt.show()
```

### Gráfico de densidade (idade)

```python
# Gráfico de densidade (idade)
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='Idade', fill=True)
plt.xlabel('Idade')
plt.ylabel('Densidade')
plt.title('Densidade da Idade')
plt.show()
```

### Gráfico de violin plot (tempo de internação por diagnóstico)

```python
# Gráfico de violin plot (tempo de internação por diagnóstico)
plt.figure(figsize=(8, 6))
sns.violinplot(x='Diagnóstico', y='Tempo de Internação', data=df)
plt.xlabel('Diagnóstico')
plt.ylabel('Tempo de Internação')
plt.title('Tempo de Internação por Diagnóstico')
plt.show()
```

### Gráfico de swarm plot (idade por gênero)

```python
# Gráfico de swarm plot (idade por gênero)
plt.figure(figsize=(8, 6))
sns.swarmplot(x='Gênero', y='Idade', data=df)
plt.xlabel('Gênero')
plt.ylabel('Idade')
plt.title('Idade por Gênero')
plt.show()
```

### Gráfico de pairplot (relação entre variáveis)

```python
# Gráfico de pairplot (relação entre variáveis)
plt.figure(figsize=(10, 6))
sns.pairplot(df, hue='Diagnóstico')
plt.title('Relação entre Variáveis')
plt.show()
```

### Gráfico de boxplot (custo por gênero)

```python
# Gráfico de boxplot (custo por gênero)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gênero', y='Custo', data=df)
plt.xlabel('Gênero')
plt.ylabel('Custo')
plt.title('Custo por Gênero')
plt.show()
```

### Gráfico de barra agrupada (tempo de internação por diagnóstico e gênero)

```python
# Gráfico de barra agrupada (tempo de internação por diagnóstico e gênero)
plt.figure(figsize=(10, 6))
sns.barplot(x='Diagnóstico', y='Tempo de Internação', hue='Gênero', data=df)
plt.xlabel('Diagnóstico')
plt.ylabel('Tempo de Internação')
plt.title('Tempo de Internação por Diagnóstico e Gênero')
plt.legend(title='Gênero')
plt.show()
```

### Gráfico de dispersão (custo por tempo de internação)

```python
# Gráfico de dispersão (custo por tempo de internação)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Tempo de Internação', y='Custo', data=df)
plt.xlabel('Tempo de Internação')
plt.ylabel('Custo')
plt.title('Custo por Tempo de Internação')
plt.show()
```

### Gráfico de barra empilhada (número de pacientes por diagnóstico e gênero)

```python
# Gráfico de barra empilhada (número de pacientes por diagnóstico e gênero)
df_count = df.groupby(['Diagnóstico', 'Gênero']).size().reset_index(name='Count')
plt.figure(figsize=(10, 6))
sns.barplot(x='Diagnóstico', y='Count', hue='Gênero', data=df_count)
plt.xlabel('Diagnóstico')
plt.ylabel('Número de Pacientes')
plt.title('Número de Pacientes por Diagnóstico e Gênero')
plt.legend(title='Gênero')
plt.show()
```

### Gráfico de countplot (número de pacientes por diagnóstico)

```python
# Gráfico de countplot (número de pacientes por diagnóstico)
plt.figure(figsize=(10, 6))
sns.countplot(x='Diagnóstico', data=df)
plt.xlabel('Diagnóstico')
plt.ylabel('Número de Pacientes')
plt.title('Número de Pacientes por Diagnóstico')
plt.show()
```
